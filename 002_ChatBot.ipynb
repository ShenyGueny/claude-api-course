```python
# load the API key from the .env file
from dotenv import load_dotenv
load_dotenv()
```

```python
# load client and the model
from anthropic import Anthropic
client = Anthropic()
model = "claude-haiku-4-5-20251001"
```

```python
# to keep track of the conversation history from the user
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

# to keep track of the conversation history from the api
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model, 
        max_tokens = 600, 
        messages = messages
    )
    return message.content[0].text
```

```python
# use a while loop to loop for user input and generate resposnes
message_history = []
user_input = input("User: ")
while user_input.lower() != "exit":
    # add user message
    add_user_message(message_history, user_input)
    # get response from claude
    response = chat(message_history)
    print("Claude: ", response)
    add_assistant_message(message_history, response)
    user_input = input("User: ")
```
