```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-5"
```

```python
# to keep track of the conversation history from the user
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

# to keep track of the conversation history from the api
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model, 
        max_tokens = 600, 
        messages = messages
    )
    return message.content[0].text

message = client.messages.create(
    model=model, 
    max_tokens = 600, 
    messages = [
    {
        "role": "user",
        "content": "What is gradient descent?"
    }
])
```

```python
# to single out the text from the response
message.content[0].text
```

```python
# to remember previous messages, you can store them in a list and pass them to the next request
message_history = []

# add in the initial user question
add_user_message(message_history, "What is gradient descent?")
message_history

# pass the list of messages into 'chat' to get an answer
answer = chat(message_history)

# add the answer to the message history
add_assistant_message(message_history, answer)

# add in another user message
add_user_message(message_history, "write me another sentence")

# check to see if api maintained history of the conversation
answer = chat(message_history)
print(answer)
```
