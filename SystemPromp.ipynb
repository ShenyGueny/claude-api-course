```python
from dotenv import load_dotenv
load_dotenv()
```

```python
# importing claude and setting up the client and model
from anthropic import Anthropic
client = Anthropic()
model = "claude-haiku-4-5-20251001"
```

```python
# setting up system prompting
system_prompt = "You are a patient and helpful math tutor. You will not directly answer math questions from the user butt give them hints to help them solve the questions."

def add_user_message(messages, user_message):
    messages.append({"role": "user", "content": user_message})
    
def add_system_message(messages, system_message):
    messages.append({"role": "assistant", "content": system_message})

# adding system prompt to the chat
def chat_with_claude(messages, system_prompt = None):

    parameters = {
        "model": model,
        "max_tokens": 600,
        "messages": messages,
    }

    if system_prompt:
        parameters["system"] = system_prompt

    response = client.messages.create(**parameters)
    

    return response.content[0].text
```

```python
messages = []
user_message = input("Ask me a math question: ")

while user_message != "exit":
    add_user_message(messages, user_message)
    response = chat_with_claude(messages, system_prompt)
    print("Claude: ", response)
    add_system_message(messages, response)
    user_message = input("Ask me a math question: ") 
```
